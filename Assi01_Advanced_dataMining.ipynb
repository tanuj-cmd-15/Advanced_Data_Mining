```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
csv_path = '/content/heart-disease.csv'
df = pd.read_csv(csv_path)
```

```python
df.describe()
```

```python
print("Attributes in the dataset:")
print(df.columns.tolist())
print("Total number of attributes:", len(df.columns))
```

```python
print("\nData types of each attribute:")
print(df.dtypes)
```

```python
age_mean = df['age'].mean()
age_median = df['age'].median()
age_mode = df['age'].mode()[0]

chol_mean = df['chol'].mean()
chol_median = df['chol'].median()
chol_mode = df['chol'].mode()[0]

print("\nStatistical Measures:")
print("Age - Mean: {:.2f}, Median: {:.2f}, Mode: {:.2f}".format(age_mean, age_median, age_mode))
print("Cholesterol - Mean: {:.2f}, Median: {:.2f}, Mode: {:.2f}".format(chol_mean, chol_median, chol_mode))

plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
sns.boxplot(y=df['age'])
plt.title("Boxplot for Age")
plt.ylabel("Age")

plt.subplot(1, 2, 2)
sns.boxplot(y=df['chol'])
plt.title("Boxplot for Cholesterol")
plt.ylabel("Cholesterol")

plt.tight_layout()
plt.show()
```

```python

age_mean = df['age'].mean()
age_median = df['age'].median()
age_mode = df['age'].mode()[0]

chol_mean = df['chol'].mean()
chol_median = df['chol'].median()
chol_mode = df['chol'].mode()[0]


plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
sns.histplot(df['age'], kde=True, bins=15, color='skyblue', label='Age Distribution')
plt.axvline(age_mean, color='red', linestyle='--', label=f'Mean: {age_mean:.2f}')
plt.axvline(age_median, color='green', linestyle='-', label=f'Median: {age_median:.2f}')
plt.axvline(age_mode, color='blue', linestyle=':', label=f'Mode: {age_mode:.2f}')
plt.title("Age Distribution with Mean, Median, and Mode")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.legend()

plt.subplot(1, 2, 2)
sns.histplot(df['chol'], kde=True, bins=15, color='lightcoral', label='Cholesterol Distribution')
plt.axvline(chol_mean, color='red', linestyle='--', label=f'Mean: {chol_mean:.2f}')
plt.axvline(chol_median, color='green', linestyle='-', label=f'Median: {chol_median:.2f}')
plt.axvline(chol_mode, color='blue', linestyle=':', label=f'Mode: {chol_mode:.2f}')
plt.title("Cholesterol Distribution with Mean, Median, and Mode")
plt.xlabel("Cholesterol")
plt.ylabel("Frequency")
plt.legend()

plt.tight_layout()
plt.show()


age_skewness = df['age'].skew()
chol_skewness = df['chol'].skew()
age_skewness, chol_skewness
```
